```python
from time import sleep
import os, zipfile
import shutil
import win32com.client
```

```python
def extrair():
    # Fonte https://stackoverflow.com/questions/31346790/
    # unzip-all-zipped-files-in-a-folder-to-that-same-folder-using-python-2-7-5

    dir_name = 'C:\\Users\\F116295\\Desktop\\Bases'
    extension = ".ZIP"

    os.chdir(dir_name)

    # change directory from working dir to dir with files

    for item in os.listdir(dir_name):
        
        # loop through items in dir
        if item.endswith(extension):
            # check for ".zip" extension
            file_name = os.path.abspath(item)
            
            # get full path of files
            zip_ref = zipfile.ZipFile(file_name)
            
            # create zipfile object
            zip_ref.extractall(dir_name)
            
            # extract file to dir
            zip_ref.close()
            
            # close file
            os.remove(file_name)

            # delete zipped file
            print("Feito: " + file_name)

    extension = ".zip"

    os.chdir(dir_name)
    
    # change directory from working dir to dir with files
    for item in os.listdir(dir_name):
        
        # loop through items in dir
        if item.endswith(extension):
            
            # check for ".zip" extension
            file_name = os.path.abspath(item)
            
            # get full path of files
            zip_ref = zipfile.ZipFile(file_name)
            
            # create zipfile object
            zip_ref.extractall(dir_name)
            
            # extract file to dir
            zip_ref.close()
            
            # close file
            os.remove(file_name)
            
            # delete zipped file
            print("Feito: " + file_name)
```

```python
def renomear_arquivos(diretorio, inicio_nome, novo_nome, extencao):
    try:
        try:
            for filename in os.listdir(diretorio):
                if filename.startswith(inicio_nome):
                    arquivo_nome_original = filename

            nome_antigo = diretorio + '\\' + arquivo_nome_original
            nome_novo = diretorio + '\\' + novo_nome + extencao

            # Renomeando
            os.rename(nome_antigo, nome_novo)

            print(nome_antigo + ' renomeado para ' + nome_novo)

        except:
            for filename in os.listdir(diretorio):
                if filename.startswith(inicio_nome.upper()):
                    arquivo_nome_original = filename

            nome_antigo_maiuscula = diretorio + '\\' + arquivo_nome_original
            nome_novo_maiuscula = diretorio + '\\' + novo_nome + extencao

            # Renomeando
            os.rename(nome_antigo_maiuscula, nome_novo_maiuscula)

            print(nome_antigo_maiuscula + ' renomeado para ' + nome_novo_maiuscula)

    except:
        pass
```

```python
def copiar_arquivos(diretorio_atual, diretorio_novo, nome, extencao):

    try:
        dir_atual = diretorio_atual + '\\' + nome + extencao
        dir_novo = diretorio_novo + '\\' + nome + extencao

        shutil.copyfile(dir_atual, dir_novo, follow_symlinks=True)

        print('Copiado para ' + dir_novo)

    except:
        pass
```
